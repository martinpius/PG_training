```python
import re
# First clean the data (both origin & destination)
ticket['origin_clean'] = (
    ticket['ORIGIN']
    .astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r'[^a-z0-9\s\-\/]', '', regex=True)
)

ticket['destination_clean'] = (
    ticket['DESTINATION']
    .astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r'[^a-z0-9\s\-\/]', '', regex=True)
)
```

```python
def assign_region(route, lookup):
    '''
    # Kamuzora ile ishu yako ndo hapa
    route: Hapa itasimama kama 'destination_clean', na 'origin_clean'
    'lookup' hapa ni ile dictionary ya mikoa haikwepeki
    '''
    if pd.isna(route): # Hapa unachek kama ile data ina missing
        return np.nan # Itarudisha missing au kama unajua ni wapi utaandika

    for keyword, region in lookup.items():
        # Uta tafuta pattern (yale majina yasiyo na mpangilio kama vijijini etc (but its cover all pattern in your column))
        pattern = rf'\b{re.escape(keyword)}\b'
        if re.search(pattern, route):
            return region
    return np.nan # Kama kwenye data yako haipo fill missing
```

```python
# Mfano utakayotengeneza (Hapa utatumia kila siku): You only build this lookup once
lookup = {
    "Dar es Salaam": ["dar", "kigamboni", "mbezi", "mbagala", "ubungo"],
    "Morogoro": ["moro", "morogoro", "msamvu", "ifakara"],
    "Arusha": ["arusha", "kisongo", "usa river"],
}
```

```python
# Flat it before you use
lookup = {
    alias: region
    for region, aliases in lookup.items()
    for alias in aliases
}
```

```python
# Apply hapa: Utapata clean data za mwisho (origin_detected, destination_detected)
```

```python
ticket['origin_detected'] = ticket['origin_clean'].apply(lambda x: assign_region(x, lookup))
```

```python
ticket['destination_detected'] = ticket['destination_clean'].apply(lambda x: assign_region(x, lookup))
```
