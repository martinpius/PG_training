```python
def absolute(a: float = 2.4)-> float:
    '''
    a: is an argument of our function
    named 'absolute"
    -------
    This function compute the absolute value
    of (a - 20) and returns its output
    '''
    rn = random.choice([30, 10, 3, 3.6])
    out = abs(a-rn)
    res = [rn, out]
    return res
```

```python
def latraCalculater(fare_per_km = 250 ,
                    distance_covered=0):
    bus_fare = fare_per_km * distance_covered
    return bus_fare
```
