```python
# Load dataset from scikit-learn [To demonstrate the rest of Day 2 we use real life dataset]
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)

# Add target variable (house value)
df['MedHouseVal'] = california.target * 100000  # Convert to dollar values

print("Dataset shape:", df.shape)
df.head(3)
```

```python
# Data inspections and cleaning
# 1. Rename columns for clarity
df = df.rename(columns={
    'MedInc': 'MedianIncome',
    'AveRooms': 'AvgRooms',
    'AveBedrms': 'AvgBedrooms',
    'AveOccup': 'AvgOccupancy'
})

# 2. Check metadata
print("Columns:", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())

# Add synthetic missing values for demonstration
df.loc[100:150, 'Population'] = np.nan
```

```python
print("\nMissing values:\n", df.isnull().sum())
```

```python
display(df.head(3))
```

```python
# Filtering and transformation
# Create categorical income groups
df['IncomeCategory'] = pd.cut(df['MedianIncome'],
                              bins=[0, 3, 6, 9, 15],
                              labels=['Low', 'Medium', 'High', 'Very High'])
```

```python

print("Merged dataset columns:", merged_df.columns.tolist()) # Save the merged data if needed
```

```python
coastal = df[df['LatCluster'].isin([2,3])]['MedHouseVal']  # Coastal clusters
inland = df[df['LatCluster'].isin([0,4])]['MedHouseVal']   # Inland clusters

# Perform two-sample t-test
t_stat, p_val = stats.ttest_ind(coastal, inland, equal_var=False)
print(f"T-statistic: {t_stat:.2f}, p-value: {p_val:.4f}")

if p_val < 0.05:
    print("Reject H₀: Significant price difference between coastal and inland properties")
else:
    print("No significant difference detected")
```

```python
# Prepare data groups---> For oneway ANOVA
income_groups = [
    df[df['IncomeCategory'] == cat]['MedHouseVal']
    for cat in ['Low', 'Medium', 'High', 'Very High']
]
```

```python
# One-way ANOVA
f_stat, p_val = stats.f_oneway(*income_groups)
print(f"F-statistic: {f_stat:.1f}, p-value: {p_val:.4f}")
```

```python
# 2. Regression Analysis
# 2.1 OLS Regression with StatsModels
```

```python
# Prepare the data for MLR

X = df[['MedianIncome', 'AvgRooms', 'HouseAge', 'Latitude']]
X = sm.add_constant(X)  # Adds intercept term
y = df['MedHouseVal'] #Price of the house

# Fit model
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Confidence intervals for the regression coeff
# Get 95% CI for coefficients
conf_int = model.conf_int(alpha=0.05)
conf_int.columns = ['2.5%', '97.5%'] # Two tailed test [L=0.05/2, R = 1-]
print("\nConfidence Intervals:")
print(conf_int)
```

```python
# Checking the Assumptions for LR

#---> Normality assumption
residuals = model.resid # Fetch the residuals
qqplot(residuals, line='45', fit=True) # Plot the residuals --> q-q plot
plt.title('Q-Q Plot of Residuals')
# Try a transformation of the target variable (e.g., log, square root) if not normal
```

```python
#---> Homoscedasticity--# Must be random above and below
plt.scatter(model.fittedvalues, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs Fitted Values')
#Apply log or square-root transformation to the dependent variable.
#Use weighted least squares (WLS).[When Violeted]
```

```python
# Train-test split (forecasting 2 days = 192 steps)
train = ts.loc[:"2017-12-31 23:45:00"]
test = ts.loc["2018-01-01 00:00:00":]
```

```python
#  Fit ARIMA model (example: p=2, d=0, q=2; tune this)
model = ARIMA(train, order=(2,0,2))
model_fit = model.fit()
print(model_fit.summary())
```

```python
# Forecasting
forecast = model_fit.forecast(steps=192)
```

```python
# Evaluation
rmse = np.sqrt(mean_squared_error(test[:192], forecast))
mae = mean_absolute_error(test[:192], forecast)
mape = np.mean(np.abs((test[:192] - forecast) / test)) * 100
```

```python
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"MAPE: {mape:.2f}%")
```

```python
# Plot forecast vs actual
plt.figure(figsize=(14,5))
plt.plot(test[:192].index, test[:192], label='Actual')
plt.plot(test[:192].index, forecast, label='Forecast')
plt.title("2-Day Ahead Forecast vs Actual")
plt.legend()
plt.grid(True)
plt.show()
```

```python
## 1. Reusable Functions for Common Tasks

def load_housing_data(file_path='california_housing_cleaned.csv'):
    """Load and preprocess data with consistent formatting"""
    df = pd.read_csv(file_path)
    #display(df.head(3))
    df['IncomeCategory'] = pd.cut(df['MedianIncome'], bins=[0, 3, 6, 9, 15],
                                  labels=['Low', 'Medium', 'High', 'Very High'])
    return df
```

```python
def save_results(results_dict):
    """Save analysis outputs to JSON"""
    with open('final_analysis.json', 'w') as f:
        json.dump(results_dict, f)
    print(f"Results saved to {'final_analysis.json'}")
```

```python
DATA_PATH = "/content/california_housing_cleaned.csv"
OUTPUT_DIR = 'content/'
```

```python
def run_full_analysis():
    """End-to-end analysis workflow-reusing the above"""
    df = load_housing_data(DATA_PATH)
    df.drop("IncomeCategory", axis = 1, inplace = True)

    # Perform analysis
    summary_stats = df.describe().to_dict()
    corr_matrix = df.corr().to_dict()

    # Save outputs
    save_results({'stats': summary_stats, 'correlations': corr_matrix})

    return summary_stats, corr_matrix
```

```python
sumsta, cor = run_full_analysis()
```

```python
# Batch processing automated:
df = pd.read_csv(DATA_PATH)

# Batch process multiple parameter combinations
param_grid = {
    'income_bins': [[0, 3, 6, 9, 15], [0, 5, 10, 15]],
    'age_thresholds': [40, 50, 60]
}

results = []

for bins in param_grid['income_bins']:
    for threshold in param_grid['age_thresholds']:
        # Automated categorization
        df['IncomeGroup'] = pd.cut(df['MedianIncome'], bins=bins)
        df['OldHouse'] = np.where(df['HouseAge'] > threshold, 1, 0)

        # Store configuration and result
        results.append({
            'bins': bins,
            'threshold': threshold,
            'old_houses_pct': df['OldHouse'].mean()
        })

pd.DataFrame(results).to_csv('parameter_sweep_results.csv', index=False)
```

```python
# Automate report
# Generate PDF report with results

def create_report(summary_stats, output_file='report.pdf'):
    """Convert analysis results to PDF"""
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    pdf.cell(200, 10, txt="California Housing Analysis Report", ln=1, align='C')

    # Add summary table
    pdf.cell(200, 10, txt="Key Statistics:", ln=1)
    for stat, value in summary_stats['MedHouseVal'].items():
        pdf.cell(200, 10, txt=f"{stat}: {value:.2f}", ln=1)

    pdf.output(output_file)

# Usage
stats = df.describe().to_dict()
create_report(stats)
```
